```python
# read in dataframe and list all kyes.  
import polars as pl
from pathlib import Path    
from pprint import pprint


datapath = Path("artifacts/2025-08-03_18-08-36_loris-mbp-cable-rcn-com/keypairs")
parquet_path = datapath / "keypairs.parquet"
csv_path = datapath / "keypairs.csv"


df_good = pl.read_csv(csv_path, infer_schema_length=159396)
df_good 
```

```python
ukeys = { key:idx for idx,key in  enumerate(sorted(set(df_good['key1'].to_list()).union(set(df_good['key2'].to_list()))))}
pprint(ukeys)
```

```python
df_good = (
    df_good.lazy()
    .with_columns(
        pl.col("key1").map_elements(lambda x: ukeys.get(x, -1), return_dtype=pl.Int64).alias("key1_mapped"),
        pl.col("key2").map_elements(lambda x: ukeys.get(x, -1), return_dtype=pl.Int64).alias("key2_mapped")
    )
    ).collect()
df_bad  = df_good.filter(pl.col("error_description") != 'No error')
df_bad
```

```python
# Check for key1_mapped or key2_mapped being -1
if df_good.filter((pl.col("key1_mapped") == -1) | (pl.col("key2_mapped") == -1)).is_empty():
    print("All keys mapped successfully.")
else:
    print("There are unmapped keys in key1_mapped or key2_mapped.")
    
# Verify all error_description are not 'No error'
if df_good.filter(pl.col("error_description") == 'No error').is_empty():
    print("All error_descriptions are valid.")
else:
    print("There are entries with 'No error' in error_description.")
```

```python
"""
user_id:
    <platform_id>_<video_id>_<session_id>:
        np.ndarray of key1_press, key2_press, key1_mapped
    )
"""
# Create a input_id from <platform_id>_<video_id>_<session_id>
df_kvc = (
    df_good.lazy()
    .with_columns(
        pl.concat_str([
            pl.col("platform_id"),
            pl.col("video_id"),
            pl.col("session_id")
        ], separator="_").alias("input_id"),
        pl.concat_list([
            pl.col("key1_press").cast(pl.Int64),
            pl.col("key1_release").cast(pl.Int64),
            pl.col("key1_mapped").cast(pl.Int64)
        ]).alias("data"),
    ).filter(pl.col("valid") == True)
).collect()

df_kvc
```

```python

import numpy as np

def get_train_val_split(df, train_platform_ids = [1, 2], hold_out_platform_id=[3]):
    train = {}
    test = {}
    for u in df_kvc["user_id"].unique().to_list():
        
        train[u] = {}
        test[u] = {}

        # Get user's data
        df_user = df_kvc.filter(pl.col("user_id") == u)
        
        assert not df_user.is_empty(), f"ERROR: No data found for user {u}: {df_user}"
        
        
        for input_id in df_user["input_id"].unique().to_list():
            df_data = df_user.filter(pl.col("input_id") == input_id)
            
            platform_ids = df_data['platform_id'].unique().to_list()
            assert len(platform_ids) == 1, f"ERROR: Multiple platform_ids found for input_id {input_id} for user {u}: {platform_ids}"
                
            arr = np.stack(df_data['data'].to_list())
            # arr = df_data['data'].to_list()
            
            if platform_ids[0] in hold_out_platform_id:
                # If the platform_id is in hold_out_platform_id, skip this input_id
                test[u][input_id] = arr
            else:
                # Otherwise, add to train
                train[u][input_id] = arr
                
    return train, test


for test_platform_id in [1,2,3]:
    hold_out_platform_id = [test_platform_id]
    train_platform_ids = {1,2,3}.difference(set(hold_out_platform_id)
                                            )
    print(f"Train on platform_ids {train_platform_ids}, hold out platform_id {hold_out_platform_id}")
    train, test = get_train_val_split(df_kvc, train_platform_ids, hold_out_platform_id)
    print(f"Train users: {len(train)}, Test users: {len(test)}")

    # save features
    version_id = "2025-08-10_20-08-33_loris-mbp-cable-rcn-com"
    sub_path = f"test_platform_{test_platform_id}"
    save_path = Path(f"artifacts") / version_id / sub_path
    save_path.mkdir(parents=True, exist_ok=True)
    
    with open(save_path / "train.npy", "wb") as f:
        np.save(f, train)
        
    with open(save_path / "test.npy", "wb") as f:
        np.save(f, test)

```
